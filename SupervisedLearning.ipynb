### Example 3.1 Python SVM Classifications

#Supervised Learning: Classifications

```python
from sklearn import svm

#Height[cm], Weight[kg], Shoesize[UK]
X = [[170, 70, 10], [180, 80,12], [170, 65, 8],[160, 55, 7]]

y = [0, 0, 1, 1] #Gender, 0: Male, 1: Female

clf = svm.SVC()
clf.fit(X, y)

#Predict
p = clf.predict([[160, 60, 7]])

print(p)
```

### Example 3.2 shows an SVM classification

```python
#Example 3.2 Python SVM Iris Classifications
from sklearn import svm, datasets

iris = datasets.load_iris()

# Take the first two features: Sepal length and Sepal width
X = iris.data[:, :2]
y = iris.target #0: Setosa, 1: Versicolour, 2:Virginica
print(y)

clf = svm.SVC()
clf.fit(X, y)

#Predict the flower for a given Sepal length and width
p = clf.predict([[5.4, 3.2]])
print(p)
```

### แบบฝึกหัด 3.2
แก้ไขโปรแกรม Python จากตัวอย่างที่ 3.2 ให้ใช้คุณลักษณะที่สามและสี่ (ความยาวและความกว้างของกลีบดอก) เป็น X แล้วเปรียบเทียบผลลัพธ์กับตัวอย่างที่ 3.2

### Example 3.3 Load CSV

```python
from sklearn import svm, datasets
import pandas as pd

df = pd.read_csv('iris.csv')

# เลือก features (คอลัมน์ 0 และ 1 คือ sepal length และ sepal width)
X = df.values[:, :2]

# เลือก label (คอลัมน์ที่ 4 คือ species ที่ถูกแทนด้วย 0 = setosa, 1 = versicolor, 2=virginica )
y = df.values[:, 4]

# สร้างและฝึกโมเดล SVM
clf = svm.SVC()
clf.fit(X, y)

# ทำนายชนิดของดอกไม้สำหรับ sepal length = 5.4 และ sepal width = 3.2
p = clf.predict([[5.4, 3.2]])
print(p)
```

###Example 3.4 import csv with URL

```python
#Example 3.4 import csv with URL
from sklearn import svm, datasets
import pandas as pd
# Import the pyplot module from matplotlib
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# นำเข้าข้อมูลจาก URL
url = "https://raw.githubusercontent.com/scikit-learn/scikit-learn/6e9039160f0dfc3153643143af4cfdca941d2045/sklearn/datasets/data/iris.csv"
df = pd.read_csv(url)

print(df.shape)
print('---------------------------------')
print(df.head(3))
print('---------------------------------')
print(df.tail(3))
print('---------------------------------')
print(df.describe())
print('---------------------------------')

#count NAN
print(df.isna().sum().sum())
print('---------------------------------')
#drop NAN values
df = df.dropna()

print(df.groupby(df.columns[4]).size())
print('---------------------------------')

# histograms
df.hist()
plt.show()
# scatter plot matrix
scatter_matrix(df)
plt.show()

# เลือก features (คอลัมน์ 0 และ 1 คือ sepal length และ sepal width)
X = df.values[:, :4]

# เลือก label (คอลัมน์ที่ 4 คือ species ที่ถูกแทนด้วย 0 = setosa, 1 = versicolor, 2=virginica )
y = df.values[:, 4]

# สร้างและฝึกโมเดล SVM
clf = svm.SVC()
clf.fit(X, y)

# ทำนายชนิดของดอกไม้สำหรับ sepal length = 5.4 และ sepal width = 3.2
p = clf.predict([[5.4, 3.2, 1.5, 0.2]])
print(p)
```
